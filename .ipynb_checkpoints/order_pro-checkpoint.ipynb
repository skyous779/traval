```python
list = []
with open('../result.txt', 'r') as f:
    for line in f:
        list.append(line.strip())
list[1]
```

```python
### 把对应的编号放入num_result中
num_result = []
for numb in list:
    p1 = numb.find('.')
    order_n = int(numb[:p1]) ## order_n = int(numb[4:p1])
    num_result.append(order_n)
    print(order_n)
```

```python
### 把对应的结果放入output中
output = []
for numb in list:
    p2 = numb.find(',')
    order_n = int(numb[p2+1:]) ## order_n = int(numb[4:p1])
    output.append(order_n)
    print(order_n)
```

```python
output[1]
```

```python
# 保存没有排序的output
with open("./result_output.txt", "w") as f:
    for item in range(len(output)):
        print(output[item])
        f.writelines(str(output[item]))
        f.writelines('\n')
    f.close()
```

```python
order_list = []
for i in range(len(num_result)):
    order_list.append((num_result[i], output[i]))
print(order_list)
```

```python
sorted_by_second = sorted(order_list, key=lambda tup: tup[0])
```

```python
sorted_by_second
```

```python
with open("./result_sort.txt", "w") as f:
    for item in sorted_by_second:
        print(item[1])
        f.writelines(str(item[1]))
        f.writelines('\n')
    f.close()
```
